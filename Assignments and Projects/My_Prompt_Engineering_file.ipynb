```python
from openai import OpenAI
from fpdf import FPDF



client = OpenAI(api_key=api_key1)

while True:
    user_input = input("Enter your Question:- ")
    if user_input.lower() == 'exit':
        break
    else:
        prompt = """Please make sure all responses are in 20 pages PDF
        with 4 diagrams and 4 tables."""

        response = client.chat.completions.create(
            model="gpt-4.1-mini",   #valid model
            messages=[
                {"role": "system", "content": "You are a helpful AI assistant."},
                {"role": "user", "content": f"{user_input}\n\n{prompt}"}
            ]
        )

        #  Get the response text
        answer = response.choices[0].message.content
        print(answer)   # still print on console

        #  Save response automatically into PDF
        pdf = FPDF()
        pdf.add_page()
        pdf.set_font("Arial", size=12)
        pdf.multi_cell(0, 10, answer)

        # File will be saved with the question as filename
        filename = user_input.replace(" ", "_") + ".pdf"
        pdf.output(filename)
        print(f" PDF saved as {filename}")
```
